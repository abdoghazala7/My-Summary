<h1 align="center">
Relational plot (relplot)  


#### Note: kind can be other things like line or other parameters

https://seaborn.pydata.org/generated/seaborn.relplot.html # link for relplot

```python
sns.relplot(x='duration',y='balance',data=df,kind='scatter')
plt.show()
```

#### what if you want to make more than 1 plot depending on columns
#### like if col smokers = yes , and col smokers = no 
#### and you want to visualize every col of them with the data

```python
sns.relplot(x='duration',y='balance',data=df,kind='scatter',col='smoker')
plt.show()
```

#### you can also replace col by row to show different alignment 

```python
sns.relplot(x='duration',y='balance',data=df,kind='scatter',row='smoker',hue='travelled before')
plt.show()
```

#### and you can use row and col tigether also


```python
sns.relplot(x='duration',y='balance',data=df,kind='scatter',col='smoker',row='travelled before')
plt.show()
```

##### now lets asume you want to visualize relation
##### between data x and y through days where day = ["day 1", "day 2", "day 3", "day 4"]

```python
sns.relplot(x="duration",y="balance",data=df,kind="scatter",col='day',col_wrap=2) ## this will show 2 plots in 1 row
plt.show()
```

```python
sns.relplot(x="duration",y="balance",data=df,kind="scatter",col='day',col_wrap=4) ## this will show 4 plots in 1 row
plt.show()
```

```python
sns.relplot(x="duration",y="balance",data=df,kind="scatter",col="day",hue='smoker',col_wrap=2,col_order=["day 4","day 3","day 2","day 1"])
plt.show()
## note here that col_wrap parameter will help to make 2 plots in each row while col_order will change the order of the day 
```

### Size 

```python
sns.scatterplot(x='duration',y='balance',data=df,size='age',hue='age')
plt.show()
## ok but what did I do here :)
## here there is a variable in the dataset which is age that I maked it as a refrence 
## for data with age high it will be get larger points
## and for the hue its a good practice to make larger points with different color from smallest
## if you didn't do the hue parameter this will end by points near each other 
## which may not help much to visualize the data
## here larger groups of points will get darker color by default
```

### Style

```python
sns.relplot(x='duration',y='balance',data=df,kind='scatter',hue='smoker',style='smoker')
plt.show()
## her based on yes, or no there will be 2 colors because of hue
## and based on yes or no for smoker there will be 2 points styles like (x , .)
```

### transparency

## I will add another data named data_for_visualizing2
### and that is because it has more continuous data 
### but I will do little cleaning on it here first

```python
df2 = pd.read_csv('data_for_visualizing2.csv')
df2
```

```python
df2.info()
```

### Note that outcome is int but It should mean True or False
### I will replace 1 by 'yes' and 0 by 'no'

```python
df2['Outcome'].replace([1, 0], ['yes','no'], inplace=True)
df2
# And this is the new data ... note that I don't care about how clean is the data here
# here I only care about visualiztions
# so I may plot relations between data that in the real world looks to have no relation together
# but as said I only care about visualization syntax
```

```python
### NOTE: IF LINE PLOT WAS GIVEN MULTIPLE VALUES ON 1 AXIS IT WILL PLOT THERE MEAN
```

```python
sns.relplot(x='Pregnancies',y='Glucose',data=df2,kind='line', style='Outcome', hue='Outcome', dashes=False)
plt.show()
# Notice the shaded area
```

### Now I will change this shaded area to be representing standard devetion then I will make no shade area
### The Parameter that do this is ci
### Note that the line there is the mean of this values

```python
sns.relplot(x='Pregnancies',y='Glucose',data=df2,kind='line', style='Outcome', hue='Outcome', dashes=False, ci='sd')
plt.show()
```

```python
sns.relplot(x='Pregnancies',y='Glucose',data=df2,kind='line', style='Outcome', hue='Outcome', dashes=False, ci=None)
plt.show()
# this will remove the shaded area
```

categorical plot (catplot)  

## count using catplot

```python
sns.catplot(x='Outcome',data=df2,kind='count')
plt.show()
```

```python
sns.catplot(x='Outcome',data=df2,kind='count', order=['no', 'yes'])
plt.show()
# I changed order
```

## barplot using catplot

```python
sns.catplot(x='BloodPressure',y='Insulin',data=df2,kind='bar')
sns.catplot(x='Outcome',y='Insulin',data=df2,kind='bar')
plt.show()
```

## boxplot using catplot

```python
sns.catplot(x='BloodPressure',y='Insulin',data=df2,kind='box')
sns.catplot(x='Outcome',y='Insulin',data=df2,kind='box')
plt.show()
```

### removing outliers:
### we use sym=''

```python
sns.catplot(x='Outcome',y='Insulin',data=df2,kind='box',sym='')
plt.show()
```

#### by defult whiskers extend to 1.5*IQR 
#### you can change it ex: whis=2.0    it makes 2*IQR
#### show 5th and 95th % whis=[5, 95]        then lower will show to 5 and upper to 95
#### if u want to show min and max of data whis = [0, 100]

```python
sns.catplot(x='Outcome',y='Insulin',data=df2,kind='box',sym='',whis=2.0)
sns.catplot(x='Outcome',y='Insulin',data=df2,kind='box',sym='',whis=[60,90])
plt.show()
# look at y-axis and you will note the differences
```

## pointplot using catplot


#### it shows mean of quantitative values
#### with 95% confidence intervals
#### the difference between point plot and line plot : that point plot is categorical
#### the difference between point plot and bar plot : that point plot helps to see slops through intervals

```python
sns.catplot(x='Outcome',y='Insulin',data=df2,kind='point')
# if you want to remove lines connecting the points make join = False
sns.catplot(x='Outcome',y='Pregnancies',data=df2,kind='point',color='red',join=False)
# one of the good things that you can even make ci using median not mean and you can ci = None
sns.catplot(x='Outcome',y='Glucose',data=df2,kind='point',color='green',ci=None)
plt.show()
```

```python
sns.catplot(x='Outcome',y='Glucose',data=df2,kind='point',estimator=np.median,capsize=0.2)
# I changed capsize here
plt.show()
```

## change style

## change palette
### you have palattes of :    
### "RdBu" or its reverse "RdBu_r"            its short for red blue
### "PRGn" or its reverse "PRGn_r"            its short for purple green


```python
sns.set_palette('PRGn')
sns.catplot(x='day',data=df,kind='count')
sns.set_palette('RdBu_r')
sns.catplot(x='day',data=df,kind='count')
plt.show()
```

### or you can make your palettes like:
my_palette = ['blue', 'cyan', 'green', 'red']


```python
my_palette = ['blue', 'cyan', 'green', 'purple']
sns.set_palette(my_palette)
sns.catplot(x='day',data=df,kind='count')
# colors I made
plt.show()
```

### how to do titles for AxesSubplot

```python
g = sns.scatterplot(x='age',y='job',data=df)
g.set_title("This is the title :)",y=1.05)
# note that y is the distance above title 
plt.show()
```

```python
g = sns.scatterplot(x='age',y='job',data=df)
g.set_title("this title has higher vertical distance :)",y=2.05)
plt.show()
```
