```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#  Ploting more than 1 variable on y-axis

```python
x = np.arange(0,10) 

y1 = np.random.randint(1,100000,10)
plt.plot(x,y1)
y2 = np.random.randint(10,100000,10)
plt.plot(x,y2)
y3 = np.random.randint(100,100000,10)
plt.plot(x,y3)

plt.show()

# If plot looks not good enough, that is because that data is randomized
```

#  adding title , linestyle and legend to the plot

```python
plt.plot(x,y1 , linestyle = '--')
plt.plot(x,y2)
plt.plot(x,y3)

plt.title('Relation between X and many Y variables')
plt.xlabel('X-axis data')
plt.ylabel('Y-axis data')

plt.legend(['y1','y2','y3'])
```

# bar plot

```python
x = np.linspace(0,20,20)
y = np.random.randint(50,size = 20)
plt.bar(x,y) 
```

### Histogram

```python
data = np.random.randn(5000)
plt.hist(data ,bins = 20 ,alpha=0.5) 
plt.show()
```

### Pie

```python
plt.pie([15,30,45,10], labels=('Egypt', 'Syria', 'Tunise', 'Morroco'), explode= [0,0,0,0.3], autopct= '%1.1f%%', shadow= True)
plt.axis('equal')
plt.show()
```

# Scatter

```python
x = np.random.randint(1,16,30)
y = np.random.randint(1,16,30)
color_array = np.random.randint(1,16,30)

plt.scatter(x,y, s=100, c=color_array)  
plt.colorbar()
plt.show()
```

# subplot

```python
plt.subplots(figsize=(20, 10))
# add_subplot(nrows, ncols, index)
plt.bar(x,y) 
plt.show()
```
