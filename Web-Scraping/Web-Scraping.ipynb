```python
import requests
from bs4 import BeautifulSoup as bs
```

# to check

```python
url = 'https://z-lib.io/fulltext/programming%20concepts?'
request = requests.get(url)
request
```

# to organize data

```python
soup = bs(request.content , 'html.parser')
soup
```

```python
contianers = soup.find_all('div' ,class_ = 'resItemBox resItemBoxBooks exactMatch')
contianers
```

# to get all 

```python
for contianer in contianers :
    
    name = contianer.find_all('h3', itemprop="name")
    name = name[0].text.strip()
    
    puplisher = contianer.find_all('a',itemprop="publisher")
    puplisher = puplisher[0].text
    
    author = contianer.find_all('div', class_="authors")
    author = author[0].text.strip()
    
    year = contianer.find_all('div' , class_="property_value")
    year = year[0].text
    
    print(f'name : {name}\npuplisher : {puplisher}\nauthor : {author}\nyear : {year}\n')    
```

## to store this data in csv file  

```python
import csv
```

```python
with open('z-library.csv' , 'w' , encoding = 'utf-8') as f :
    obj = csv.writer(f)
    obj.writerow(['name' , 'puplisher' , 'author' , 'year'])
    
    for contianer in contianers :
    
        name = contianer.find_all('h3', itemprop="name")
        name = name[0].text.strip()

        puplisher = contianer.find_all('a',itemprop="publisher")
        puplisher = puplisher[0].text

        author = contianer.find_all('div', class_="authors")
        author = author[0].text.strip()

        year = contianer.find_all('div' , class_="property_value")
        year = year[0].text
        
        obj.writerow([name , puplisher , author , year])
        
print('done.....')
```

#  Scraping for more pages 

```python
with open('z-library.csv' , 'w' , encoding = 'utf-8') as f :
    obj = csv.writer(f)
    obj.writerow(['name' , 'puplisher' , 'author' , 'year'])
    
    for i in range(1,6) :
        request = requests.get(f'https://z-lib.io/fulltext/programming%20concepts?page={i}') 
        soup = bs(request.content , 'html.parser')
        contianers = soup.find_all('div' ,class_ = 'resItemBox resItemBoxBooks exactMatch')
        
        for contianer in contianers :
    
            name = contianer.find_all('h3', itemprop="name")
            name = name[0].text.strip()
            
            try :
                puplisher = contianer.find_all('a',itemprop="publisher")
                puplisher = puplisher[0].text
            except :
                puplisher = 'Nan'
                
            author = contianer.find_all('div', class_="authors")
            author = author[0].text.strip()

            year = contianer.find_all('div' , class_="property_value")
            year = year[0].text

            obj.writerow([name , puplisher , author , year])
```
